```python
import sympy as sp
import numpy as np
```

```python
t = sp.symbols('t')
P11, P12, P22 = sp.symbols('P11 P12 P22')
P = sp.Matrix([[P11, P12], [P12, P22]])
A = sp.Matrix([[0, 1], [-2, 2]])
B = sp.Matrix([[0], [2]])
Q = sp.Matrix([[4, 3], [3, 4]])
R = 1
P_dot = -A.T*P - P*A + P*B*1*B.T*P - Q
P_dot
```

```python
# Runge-Kutta 4th order

def rk4(f, x0, t0, tf, h):
 t = np.arange(t0, tf, h)
 x = np.zeros((len(t), len(x0)))
 x[0] = x0
 for i in range(len(t)-1):
   k1 = f(x[i], t[i])
   print(x[i]+h*k1/2)
   k2 = f(x[i] + h*k1/2, t[i] + h/2)
   k3 = f(x[i] + h*k2/2, t[i] + h/2)
   k4 = f(x[i] + h*k3, t[i] + h)
   x[i+1] = x[i] + h*(k1 + 2*k2 + 2*k3 + k4)/6
 return x
```

```python
def f(x, t):
  print(x[1])
  print(x[2])
  return np.array([4*x[1]*2 + 4*x[1] - 4, -x[0] + 4*x[1]*x[2] - 2*x[1] + 2*x[2] - 3, -2*x[1]**2 + 4*x[2]**2 - 4*x[2] - 4])
```

```python
x0 = np.array([1, 1, 2])
x = rk4(f, x0, 0, 6, 0.001)

x.shape
```

```python
import matplotlib.pyplot as plt
t = np.arange(0, 6, 0.001)
t.shape
```

```python
# P11(t) vs t
plt.plot(t, x[:,0])
plt.title('P11(t) VS time')
plt.xlabel('t')
plt.ylabel('P11(t)')
plt.show()
```

```python
# P12(t) vs t
plt.plot(t, x[:,1])
plt.title('P12(t) VS time')
plt.xlabel('t')
plt.ylabel('P12(t)')
plt.show()
```

```python
# P22(t) vs t
plt.plot(t, x[:,2])
plt.title('P22(t) VS time')
plt.xlabel('t')
plt.ylabel('P22(t)')
plt.show()
```
